## Array math operations

```python
arr_1 = np.array([2, 4, 6])
arr_2 = np.array([1, 3, 5])

addition = arr_1 + arr_2
substraction = arr_1 - arr_2
multiplication = arr_1 * arr_2

print(addition)
print(substraction)
print(multiplication)
```

```python
# multiplying vector with a scalar (broadcasting)


vector = np.array([1,2])

vector * 1.6
```

## Stacking

```python
# join two or more arrays, either horizontally or vertically

a1 = np.array([[1,1],
              [2,2]])

a2 = np.array([[3,3],
               [4,4]])

print(a1)

print(a2)
```

```python
# stack vertically

vert_stack = np.vstack((a1, a2))
vert_stack
```

```python
horz_stack = np.hstack((a1, a2))
horz_stack
```
