## Introduction to the numpy.linalg sub-libary

```python
import numpy as np
```

```python
A = np.array([[4,-3,1],
              [2, 1, 3],
              [-1,2,-5]])

b = np.array([-10, 0, 17], dtype=np.dtype(float))

print(f'Matrix: ', A)
print('-----------')
print(b)
```

```python
np.shape(A)
```

```python
# solving the system

x = np.linalg.solve(A, b)
x
```

```python
# determinant

d = np.linalg.det(A)
d
```

```python
# if the system has no solution

A_2 = np.array([[1,1,1],
          [0, 1,-3,],
          [2,1,5]])

b_2 = np.array([2,1,0])

print(A_2)
print(b_2)
```

```python
print(np.linalg.solve(A_2, b_2))
```

```python
d_2 = np.linalg.det(A_2)

print(d_2)
```
